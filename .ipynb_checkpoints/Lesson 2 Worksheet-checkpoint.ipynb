# We are now going to consider the equation of the line that models the data

```python
USdata = data[data["Entity"]  == "United States"]
sns.set_style('whitegrid')
sns.lmplot(x ='Year', y ='Life expectancy', data = USdata)
```

# Consider the Equation for the line that fits the data bellow

```python
model = LinearRegression().fit(np.array((USdata["Year"])).reshape((-1, 1)), list(USdata["Life expectancy"]))
b = str(float(model.intercept_))

m=str(model.coef_)

print("y = "+m+"x + "+ b)

```
