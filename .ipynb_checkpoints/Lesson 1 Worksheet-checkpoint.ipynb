```python
country = "United States"
sns.relplot(data=data[data["Entity"]  == country], x="Year", y="Life expectancy").set(title= country+'\'s Life Expectancy by Year')
```

# Question 2:
### Condider a chart below that shows the change in life expectancy year by year in the United States. Which two years have the largest drop in life exectancy? (***Hint the number will be negative as this indicates a drop)

### Answer: (double click here):

________________
________________

```python
pd.set_option('display.max_rows', None)
expchange = data[data["Entity"]  == country][["Life expectancy" ]].rolling(window=2).apply(lambda x: x.iloc[1] - x.iloc[0])
expchange["year"] = data[data["Entity"]  == country]["Year"]
expchange.style.set_caption("Change in Life Expectancy by Year in the United States")
```
