```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import seaborn as sns
from sklearn.linear_model import LinearRegression
import os
```

```python
infantmort = pd.read_csv(os.getcwd()+"/thedata/child-deaths-igme-data.csv")
usinfantmort =infantmort[infantmort["Entity"]  == "United States"]

gdp = pd.read_csv(os.getcwd()+"/thedata/us-gdp-per-capita-maddison-2020.csv")
usgdp = gdp[gdp["Entity"] == "United States"]

womenemployed = pd.read_csv(os.getcwd()+r"/thedata/female-labor-force-participation-OECD.csv") 
uswomenemployed = womenemployed[womenemployed["Entity"] == "United States"]

marriagerate = pd.read_csv(os.getcwd()+r"/thedata/marriage-rate-per-1000-inhabitants.csv") 
usmarriagerate = marriagerate[marriagerate["Entity"] == "United States"]

schoolrev = pd.read_csv(os.getcwd()+r"/thedata/revenues-for-public-schools-by-source-gdp-us.csv") 
usschoolrev = schoolrev[schoolrev["Entity"] == "United States"]


birthoutsidemarriage = pd.read_csv(os.getcwd()+r"/thedata/share-of-births-outside-marriage.csv") 
usbirthoutsidemarriage = birthoutsidemarriage[birthoutsidemarriage["Entity"] == "United States"]

fertrate = pd.read_csv(os.getcwd()+r"/thedata/fertilityrate-children-per-woman-UN.csv") 
usfertrate = fertrate[fertrate["Entity"] == "United States"]

bigdata = pd.merge(pd.merge(pd.merge(pd.merge(pd.merge(pd.merge(usinfantmort,usgdp, on = "Year", how = "outer"),uswomenemployed, on ="Year", how = "outer"),usmarriagerate, on = "Year", how = "outer"), usschoolrev, on = "Year", how = "outer"),usbirthoutsidemarriage, on = "Year", how = "outer"), usfertrate, on = "Year", how = "outer")

bigdata = bigdata[["Year","Number of under-five deaths",
                  "GDP per capita",
                  "Female labor force participation rate (OWID based on OECD 2017 and others)",
                  "Crude marriage rate (per 1,000 inhabitants)",
                  "Local (US Bureau of the Census and NCES (2017))",
                  "State (US Bureau of the Census and NCES (2017))",
                  "Federal/Unassigned (US Bureau of the Census and NCES (2017))",
                  "Births outside of marriage (% of all births)",
                  "Fertility rate - Sex: all - Age: all - Variant: estimates"]]
```

# We will now implement this with our own datasets.


```python
sns.relplot(bigdata, x="Year", y="GDP per capita").set(title= "US GDP vs Time")


```

# We now take our y data and logarthimically transform it

```python
y = np.log(bigdata["GDP per capita"])
x = np.array(bigdata["Year"])
#newdp = pd.DataFrame(y,x)
z = pd.DataFrame()
z["Year"] = x
z["GDP per capita"] = y
sns.relplot(z, x="Year", y="GDP per capita").set(title= " Log of US GDP vs time ")
```

```python
sns.set_style('whitegrid')
sns.lmplot(x ='Year', y ='GDP per capita', data = z)
p = z.dropna()
model = LinearRegression().fit(np.array((p["Year"])).reshape((-1, 1)), list(p["GDP per capita"]))
b = str(float(model.intercept_))

m=str(model.coef_)

print("y = "+m+"x + "+ b)
```
