```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import seaborn as sns
```

```python
data = pd.read_csv("life-expectancy.csv")
```

```python
themeans = pd.melt(data, id_vars = ["Year"], value_vars = ["Life expectancy"]).groupby('Year').mean()
AveLifeExpectancy = sns.relplot(data=themeans, x="Year", y="value").set(title=' Average Life Expectancy by Year')
```

# Watch what happens when we logarithmically tranform the data:

```python
logtrans = themeans
loggeddata = themeans.value.apply(lambda x: np.log(x))

logtrans["Life expectancy"] = loggeddata
logtrans["year"] = list(themeans.index)
```

```python
themeans = pd.melt(logtrans, id_vars = ["year"], value_vars = ["Life expectancy"]).groupby('year').mean()
AveLifeExpectancy = sns.relplot(data=themeans, x="year", y="value").set(title=' Average Life Expectancy by Year')
```
